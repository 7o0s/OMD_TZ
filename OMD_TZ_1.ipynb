```python
import requests
from bs4 import BeautifulSoup as bs
import json
import re

import os
from fake_useragent import UserAgent
from selenium_stealth import stealth
import time

from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
```

```python
def get_random_chrome_user_agent(): #  Функция для генерации случайного user агента
    user_agent = UserAgent(browsers='chrome', os='windows', platforms='pc')
    return user_agent.random


def create_driver(user_id): #  Функция для создания driver selenium (Аргументы: id пользователя для конкретного сайта)
    options = Options()
    
    options.add_experimental_option("excludeSwitches", ["enable-automation"]) # Опция, исключающая переключатель "enable-automation"
    options.add_experimental_option('useAutomationExtension', False) # Опция, отключающая расширение автоматизации в браузере
    options.add_argument("start-maximized") # Аргумент для запуска браузера в режиме максимального окна
    
    script_dir = os.getcwd() # Получение пути к корневому репозиторию
    base_directory = os.path.join(script_dir, 'users') # Создание пути к репозиторию users
    user_directory = os.path.join(base_directory, f'user_{user_id}') # Создание пути к репозиторию пользователя по номеру id

    options.add_argument(f'user-data-dir={user_directory}') # Аргумент, указывающий на директорию с данными пользователя
    # options.add_argument('--disable-gpu')
    # options.add_argument('--disable-dev-shm-usage')
    options.add_argument("--disable-notifications") # Отключаются уведомления в браузере
    options.add_argument("--disable-popup-blocking") # Отключается блокировка всплывающих окон
    # options.add_argument('--no-sandbox')

    driver = webdriver.Chrome(options=options) # Создается экземпляр webdriver для Chrome с заданными параметрами options
    ua = get_random_chrome_user_agent() # Вызов функции get_random_chrome_user_agent
    stealth(driver=driver, # Вызывает функцию stealth для маскировки использования webdriver, с установкой различных параметров 
            user_agent=ua,
            languages=["ru-RU", "ru"],
            vendor="Google Inc.",
            platform="Win32",
            webgl_vendor="Intel Inc.",
            renderer="Intel Iris OpenGL Engine",
            fix_hairline=True,
            run_on_insecure_origins=True
            )
    return driver


def data_parse(driver): # Функция для парсинга данных (Аргументы: объект webdriver)
    result_list = [] # Создание пустого списка

    action = ActionChains(driver) # Вызывается функция ActionChains для инициализации action
    

    place_1_element = driver.find_element("xpath", '//div[@data-testid="BannersCarousel"]')
    action.move_to_element(place_1_element).perform()
    driver.execute_script("arguments[0].scrollTo(arguments[0].scrollWidth, 0);", place_1_element)
    time.sleep(2)
    driver.execute_script("arguments[0].scrollTo(arguments[0].scrollWidth, 0);", place_1_element)

    # Далее идут похожие блоки кода, на которых сначала курсор переходит на определенный элемент 
    # После чего происходит сбор данных с помощью поиска по тегам и их Xpath
    
    place_1 = driver.find_elements("xpath", '//div[@data-testid="BannersCarousel"]/div/div[@data-testid="advContainer"]')
    list_place_1 = []
    count_1 = 0
    for place_1_ in place_1:
        count_1 += 1
        place_1_img = place_1_.find_element("xpath", './/div/picture/img')
        place_1_image_url = place_1_img.get_attribute('src')
        place_1_meta = place_1_img.get_attribute('alt')
        place_1_content = place_1_.find_element("xpath", './/a')
        place_1_content_url = place_1_content.get_attribute('href')
        dic_1 = {'image_url': place_1_image_url, 'content_url': place_1_content_url, 'meta': place_1_meta, 'place': 1, 'position': count_1}
        list_place_1.append(dic_1)

    place_3_element = driver.find_element("xpath", '//div[@data-testid="Banners"]')
    action.move_to_element(place_3_element).perform()
    driver.execute_script("arguments[0].scrollTo(arguments[0].scrollWidth, 0);", place_3_element)

    time.sleep(2)
    
    place_2 = driver.find_elements("xpath", '//section[@data-testid="tileUnderMainCarouselBlock"]/div/ul/li[@data-testid="advContainer"]')
    list_place_2 = []
    count_2 = 0
    for place_2_ in place_2:
        count_2 += 1
        place_2_img = place_2_.find_element("xpath", './/div/picture/img')
        place_2_image_url = place_2_img.get_attribute('src')
        place_2_meta = place_2_img.get_attribute('alt')
        place_2_content = place_2_.find_element("xpath", './/a')
        place_2_content_url = place_2_content.get_attribute('href')
        dic_2 = {'image_url': place_2_image_url, 'content_url': place_2_content_url, 'meta': place_2_meta, 'place': 2, 'position': count_2}
        list_place_2.append(dic_2)

    place_3 = driver.find_elements("xpath", '//div[@data-testid="Banners"]/div/li[@data-testid="advContainer"]')
    list_place_3 = []
    count_3 = 0
    for place_3_ in place_3:
        count_3 += 1
        place_3_img = place_3_.find_element("xpath", './/div/picture/img')
        place_3_image_url = place_3_img.get_attribute('src')
        place_3_meta = place_3_img.get_attribute('alt')
        place_3_content = place_3_.find_element("xpath", './/a')
        place_3_content_url = place_3_content.get_attribute('href')
        dic_3 = {'image_url': place_3_image_url, 'content_url': place_3_content_url, 'meta': place_3_meta, 'place': 3, 'position': count_3}
        list_place_3.append(dic_3)

    place_4_elements = driver.find_elements("xpath", '//div[@data-testid="Products"]')
    for place_4_element in place_4_elements:
        action.move_to_element(place_4_element).perform()
        driver.execute_script("arguments[0].scrollTo(arguments[0].scrollWidth, 0);", place_4_element)

    time.sleep(2)

    place_4 = driver.find_elements("xpath", '//section[@id[starts-with(., "product")]]')
    list_place_4 = []
    count_4 = 0
    for place_4_ in place_4:
        count_4 += 1
        place_4_img = place_4_.find_element("xpath", './/div/div/picture/img')
        place_4_image_url = place_4_img.get_attribute('src')
        place_4_meta = place_4_img.get_attribute('alt')
        place_4_content = place_4_.find_element("xpath", './/a')
        place_4_content_url = place_4_content.get_attribute('href')
        dic_4 = {'image_url': place_4_image_url, 'content_url': place_4_content_url, 'meta': place_4_meta, 'place': 4, 'position': count_4}
        list_place_4.append(dic_4)

    place_5 = driver.find_elements("xpath", '//header[@role="banner"]/div[@data-testid="advContainer"]')
    list_place_5 = []
    count_5 = 0
    for place_5_ in place_5:
        count_5 += 1
        place_5_img = place_5_.find_element("xpath", './/div/div/div[1]')
        place_5_image_style = place_5_img.get_attribute('style')
        place_5_image_url = re.search(r'url\("([^"]*)"', place_5_image_style)
        place_5_content = place_5_.find_element("xpath", './/a')
        place_5_content_url = place_5_content.get_attribute('href')
        dic_5 = {'image_url': place_5_image_url.group(1), 'content_url': place_5_content_url, 'meta': "-", 'place': 5, 'position': count_5}
        list_place_5.append(dic_5)

    # Сохранение всех созданных списков в итоговый
    
    result_list.extend(list_place_1)
    result_list.extend(list_place_2)
    result_list.extend(list_place_3)
    result_list.extend(list_place_4)
    result_list.extend(list_place_5)

    return result_list


def create_result_json(user_id, url): # Функция для сохранения итогового списка в json файл (Аргументы: id пользователя, ссылка на сайт)
    driver = create_driver(user_id) # Вызов функции create_driver
    driver.get(url) 
    ipynb_dir = os.getcwd() # Получение пути к корневому репозиторию
    result_list = data_parse(driver) # Вызов функции data_parse
    with open(f"{ipynb_dir}\DetMir.json", "w", encoding='utf-8') as outfile:
            outfile.write(json.dumps(result_list, ensure_ascii=False, indent=4)) # Сохранение файла DetMir.json в корневой папке с словарем
```

```python
user_id = 1 # Инициализация переменной id пользователя
url = "https://www.detmir.ru/" # Ссылка на сайт
create_result_json(user_id, url) # Вызов функции create_result_json
```
